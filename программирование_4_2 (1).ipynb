```python
"""
Лабораторная работа 4.2 — Вариант 28: Издательство
Наследование, Полиморфизм, Инкапсуляция
"""

# ======================================================
# БАЗОВЫЙ КЛАСС 1 — Публикация
# ======================================================

class Publication:
    """Любая публикация издательства: книга или журнал."""

    def __init__(self, title, circulation, price):

        if circulation <= 0:
            raise ValueError("Тираж должен быть больше нуля!")

        if price <= 0:
            raise ValueError("Цена должна быть больше нуля!")

        self.title = title
        self.circulation = circulation
        self.price = price

    def calculate_royalties(self):
        """
        Полиморфный метод.
        Должен быть переопределён в дочерних классах.
        """
        raise NotImplementedError(
            "Метод calculate_royalties() должен быть реализован в дочернем классе!"
        )

    def __str__(self):
        return (
            f"«{self.title}» | "
            f"тираж: {self.circulation} шт. | "
            f"цена: {self.price} руб."
        )


# ======================================================
# ДОЧЕРНИЙ КЛАСС — Книга
# ======================================================

class Book(Publication):
    """Книга с указанием жанра."""

    def __init__(self, title, circulation, price, genre):
        super().__init__(title, circulation, price)
        self.genre = genre

    def calculate_royalties(self):
        return self.circulation * self.price * 0.15

    def __str__(self):
        return f"Книга {super().__str__()} | жанр: {self.genre}"


# ======================================================
# ДОЧЕРНИЙ КЛАСС — Журнал
# ======================================================

class Magazine(Publication):
    """Журнал с указанием периодичности выхода."""

    def __init__(self, title, circulation, price, frequency):
        super().__init__(title, circulation, price)
        self.frequency = frequency

    def calculate_royalties(self):
        return self.circulation * self.price * 0.08

    def __str__(self):
        return f"Журнал {super().__str__()} | выход: {self.frequency}"


# ======================================================
# БАЗОВЫЙ КЛАСС 2 — Автор
# ======================================================

class Author:
    """Автор издательства."""

    def __init__(self, name, contract_amount):

        if contract_amount < 0:
            raise ValueError("Сумма контракта не может быть отрицательной!")

        self.name = name
        self.contract_amount = contract_amount

    def __str__(self):
        return (
            f"Автор: {self.name} | "
            f"контракт: {self.contract_amount:,} руб."
        )


# ======================================================
# ДОЧЕРНИЙ КЛАСС — Известный автор
# ======================================================

class FamousAuthor(Author):
    """Автор, которому был выплачен аванс."""

    def __init__(self, name, contract_amount, advance):

        super().__init__(name, contract_amount)

        if advance < 0:
            raise ValueError("Аванс не может быть отрицательным!")

        self.advance = advance

    def __str__(self):
        return (
            f"Известный автор: {self.name} | "
            f"контракт: {self.contract_amount:,} руб. | "
            f"аванс: {self.advance:,} руб."
        )


# ======================================================
# КЛАСС-КОНТЕЙНЕР
# ======================================================

class PublishingPlan:
    """
    Издательский план.
    Связывает автора со списком публикаций.
    """

    def __init__(self, author):
        self.author = author
        self.items = []

    def add_item(self, publication):

        if not isinstance(publication, Publication):
            raise TypeError(
                "В план можно добавлять только объекты Publication!"
            )

        self.items.append(publication)

    def calculate_total(self):

        total = sum(
            item.calculate_royalties()
            for item in self.items
        )

        if isinstance(self.author, FamousAuthor):
            total -= self.author.advance

        return total


# ======================================================
# Точка входа в программу
# ======================================================

if __name__ == "__main__":

    try:

        print("=" * 52)
        print("   ЛР 4.2 — Издательство, Вариант 28")
        print("=" * 52)

        # ------------------------------------------------
        # Авторы
        # ------------------------------------------------

        print("\n--- Авторы ---\n")

        author1 = Author(
            "Андрей Белов",
            200_000
        )

        author2 = FamousAuthor(
            "Мария Королёва",
            500_000,
            50_000
        )

        print(author1)
        print(author2)

        # ------------------------------------------------
        # Публикации
        # ------------------------------------------------

        print("\n--- Публикации ---\n")

        book1 = Book(
            "Путь к вершинам Python",
            10000,
            799,
            "Учебник"
        )

        book2 = Book(
            "Тёмный лес",
            30000,
            499,
            "Роман"
        )

        magazine1 = Magazine(
            "Мир программирования",
            5000,
            350,
            "ежемесячный"
        )

        magazine2 = Magazine(
            "Литературный вестник",
            15000,
            250,
            "еженедельный"
        )

        print(book1)
        print(book2)
        print(magazine1)
        print(magazine2)

        # ------------------------------------------------
        # Полиморфизм
        # ------------------------------------------------

        print("\n--- Полиморфизм: calculate_royalties() ---\n")

        publications = [
            book1,
            book2,
            magazine1,
            magazine2
        ]

        for pub in publications:
            print(
                f"{type(pub).__name__:10}"
                f" «{pub.title}» → "
                f"{pub.calculate_royalties():,.0f} руб."
            )

        # ------------------------------------------------
        # Планы издания
        # ------------------------------------------------

        print("\n--- Планы издания ---\n")

        plan1 = PublishingPlan(author1)

        plan1.add_item(book1)
        plan1.add_item(magazine1)

        plan2 = PublishingPlan(author2)

        plan2.add_item(book2)
        plan2.add_item(magazine1)
        plan2.add_item(magazine2)

        # ------------------------------------------------
        # Отчёт по первому автору
        # ------------------------------------------------

        print("\n--- Отчёт: план обычного автора ---\n")

        print(f"Автор: {plan1.author.name}")

        for item in plan1.items:
            print(
                f"• {item.title:35}"
                f" гонорар: {item.calculate_royalties():>12,.0f} руб."
            )

        print(
            f"\nИТОГО К ВЫПЛАТЕ: "
            f"{plan1.calculate_total():,.0f} руб."
        )

        # ------------------------------------------------
        # Отчёт по известному автору
        # ------------------------------------------------

        print("\n--- Отчёт: план известного автора ---\n")

        print(
            f"Автор: {plan2.author.name} "
            f"(аванс {plan2.author.advance:,} руб. будет засчитан)"
        )

        for item in plan2.items:
            print(
                f"• {item.title:35}"
                f" гонорар: {item.calculate_royalties():>12,.0f} руб."
            )

        total_without_advance = sum(
            item.calculate_royalties()
            for item in plan2.items
        )

        print(
            f"\nСумма без вычета аванса: "
            f"{total_without_advance:,.0f} руб."
        )

        print(
            f"Засчитывается аванс: "
            f"-{plan2.author.advance:,.0f} руб."
        )

        print(
            f"\nИТОГО К ВЫПЛАТЕ: "
            f"{plan2.calculate_total():,.0f} руб."
        )
        # ------------------------------------------------
        # Демонстрация исключений
        # ------------------------------------------------

        print("\n--- Демонстрация исключений ---\n")

        # Ошибка: отрицательный тираж
        try:
            Book(
                "Ошибочная книга",
                -1000,
                500,
                "Тест"
            )
        except ValueError as error:
            print("ValueError:", error)

        # Ошибка: отрицательная сумма контракта
        try:
            Author(
                "Тестовый автор",
                -50000
            )
        except ValueError as error:
            print("ValueError:", error)

        # Ошибка: добавление не Publication
        try:
            plan1.add_item("Это не публикация")
        except TypeError as error:
            print("TypeError:", error)

        # Ошибка: вызов метода базового класса
        try:
            publication = Publication(
                "Базовая публикация",
                1000,
                100
            )

            publication.calculate_royalties()

        except NotImplementedError as error:
            print("NotImplementedError:", error)

    except ValueError as error:
        print("\nОшибка значения:")
        print(error)

    except TypeError as error:
        print("\nОшибка типа:")
        print(error)

    except Exception as error:
        print("\nНеожиданная ошибка:")
        print(error)
```
